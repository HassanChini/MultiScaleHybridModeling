# Description

```python
#     Aikman & Scaife model
#  mean plant weight as a function of day-degrees (DD) for different densities d=180, d= 480, d=1920, d=7680
#  
#    
#
#
#
#   In conclusion, I have plot the four denisties in one figure using subplot
#
```

# d = 180

```python
%reset -f
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
%matplotlib
plt.ion()
plt.figure()



DD=np.array([[200, 290, 335, 390],[ 180, 340, 405, 455],[ 135, 205, 315, 420 ]])
time=np.array([[13,19,22,27],[19,34,42,50],[16,23,35,43]])
intervale=np.linspace(100,550,1000)
density=np.array([180,480,1920,7680])
date=['03/09/81','14/10/81','16/03/82']




def Aikman(DD, f0, KF, r):
    return np.log( 1 + ((f0/(1-f0)) * np.exp(r*DD)) ) / ( density[0] * KF)




weights=np.array([  
                    [0.009687573, 0.044685234, 0.094008898, 0.227650969],
                    [0.004384217, 0.031238793, 0.061785221, 0.146166469],
                    [0.003410082, 0.014033932, 0.166518499, 0.791899893]])

for i in range(0,3):
    
    def objective(params):
        f0, KF, r = params
        y_pred = np.array([Aikman(DD[i][j], f0, KF, r) for j in range(4)]) 
        y_obs = weights[i]
        return sum((y_obs - y_pred) ** 2)


    params0 = [0.0005, 0.0008, 0.0005]
    result = minimize(objective, params0)
    print("Paramètres optimaux : f0 = %.8f, KF = %.8f, r = %.8f" % tuple(result.x))

    y_obs = weights[i]

    # Calcul de la somme des carrés totale
    y_mean = np.mean(y_obs)
    SST = sum((y_obs - y_mean) ** 2)

    # Calcul de la somme des carrés résiduels
    f0_opt, KF_opt, r_opt= result.x
    y_pred = np.array([Aikman(DD[i][j], f0_opt, KF_opt, r_opt) for j in range(4)])
    SSR = sum((y_obs - y_pred) ** 2)

    # Calcul du coefficient de détermination
    R2 = 1 - (SSR / SST)
    print("Coefficient de détermination : R2 = %.4f" % R2)

    plt.plot(intervale, Aikman(intervale, f0_opt, KF_opt, r_opt),ls="-",label=('date=%s'%date[i]),lw=1)
    plt.scatter(DD[i], y_obs)
    plt.xlabel('day-degrees')
    plt.ylabel('mean plant weight')
    plt.title('density=180')
    plt.legend()
plt.show()
```
