```python
%reset -f
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
%matplotlib
plt.ion()
plt.figure()



DD=np.array([200, 290, 335, 390, 180, 340, 405, 455, 135, 205, 315, 420 ])
R=np.array([54, 68.5125, 76.16842105, 95.27142857,32.4, 43.71428571, 50.38676471, 59.76486486,53.46, 65.6, 150.255, 189])
time=np.array([13,19,22,27,19,34,42,50,16,23,35,43])
intervale=np.linspace(90,350,1000)
density=np.array([180,480,1920,7680])
date=['03/09/81','14/10/81','16/03/82']


def Aikman_1(DD, R, f0, f, KF, r, n):
    EDD = 1 / ( (1/DD)+f*(1/R) )
    return np.log( 1 + ((f0/(1-f0)) * np.exp(r*EDD)) ) / ( density[0] * KF)

def Aikman_2(t, f0, KF, r):
        return np.log( 1 + ((f0/(1-f0)) * np.exp(r*t)) ) / ( density[0] * KF)


weights=np.array([0.009687573, 0.044685234, 0.094008898, 0.227650969,
                    0.004384217, 0.031238793, 0.061785221, 0.146166469,
                    0.003410082, 0.014033932, 0.166518499, 0.791899893])

def objective(params):
        f0, f, KF, r = params
        y_pred = np.array([Aikman_1(DD[i], R[i], f0, f, KF, r, time[i]) for i in range(len(R))]) 
        y_obs = weights
        return sum((y_obs - y_pred) ** 2)

params0 = [0.0005, 0.09, 0.008, 0.002]
result = minimize(objective, params0)
print("Paramètres optimaux : f0 = %.8f, f=%.8f, KF = %.8f, r = %.8f" % tuple(result.x))

y_obs = weights

# Calcul de la somme des carrés totale
y_mean = np.mean(y_obs)
SST = sum((y_obs - y_mean) ** 2)

# Calcul de la somme des carrés résiduels
f0_opt, f_opt, KF_opt, r_opt= result.x
y_pred = np.array([Aikman_1(DD[i], R[i], f0_opt, f_opt, KF_opt, r_opt, time[i]) for i in range(len(R))])
SSR = sum((y_obs - y_pred) ** 2)

# Calcul du coefficient de détermination
R2 = 1 - (SSR / SST)
print("Coefficient de détermination : R2 = %.4f" % R2)

 
EDD=1/ ((1/DD)+f_opt*(1/R))


DD=np.array([[200, 290, 335, 390],[ 180, 340, 405, 455],[ 135, 205, 315, 420 ]])
R=np.array([
            [54, 68.5125, 76.16842105, 95.27142857],
            [32.4, 43.71428571, 50.38676471, 59.76486486],
            [53.46, 65.6, 150.255, 189]])
time=np.array([[13,19,22,27],[19,34,42,50],[16,23,35,43]])
intervale=np.linspace(90,350,1000)
density=np.array([180,480,1920,7680])
date=['03/09/81','14/10/81','16/03/82']



weights=np.array([[0.009687573, 0.044685234, 0.094008898, 0.227650969],
                    [0.004384217, 0.031238793, 0.061785221, 0.146166469],
                    [0.003410082, 0.014033932, 0.166518499, 0.791899893]])


EDD=1/ ((1/DD)+f_opt*(1/R))
plt.xlabel('EDD')
plt.ylabel('mean dry weight per plant')
for i in range(0,3):
    plt.plot()
    plt.scatter(EDD[i],weights[i],label=date[i])
plt.ylabel('mean dry weight per plant') 
plt.plot(intervale,Aikman_2(intervale, f0_opt, KF_opt, r_opt),label='fitted data')
plt.title('density=180 (R2=%.4f)'%R2)
plt.legend()

```
